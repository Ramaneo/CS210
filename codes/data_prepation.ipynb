# Variable Analysis

```python
def bar_plot(variable):
    var = data[variable]
    varValue = var.value_counts()
    
    plt.figure(figsize = (9,3))
    plt.bar(varValue.index, varValue)
    plt.xticks(varValue.index, varValue.index.values)
    plt.ylabel('Frequency')
    plt.title(variable)
    plt.show()
    print("{}: \n {}".format(variable,varValue))
```

```python
categories = ['Cinsiyet','Yas','Bolge','Egitim','parti']
for c in categories:
    bar_plot(c)
```

```python
def pie_plt(column_name,datas,data_name):
    x_temp = datas[column_name].value_counts()
    labels = datas[column_name].value_counts().index
    colors = sns.color_palette('pastel')[0:5]
    fig,ax = plt.subplots(figsize=(7,7))
    g = plt.pie(x=x_temp,data=datas,labels = labels, autopct = "%1.1f%%")
    ax.set_title(data_name + " -- "+ column_name)    
    plt.show()
```

# Mapping Data

```python
# Mapping Cinsiyet
data.Cinsiyet=[1 if each=='Erkek' else 0 for each in data.Cinsiyet]

# Mapping questions
data.soru1=[1 if each=='Evet' else 0 for each in data.soru1]
data.soru2=[1 if each=='Evet' else 0 for each in data.soru2]
data.soru3=[1 if each=='Evet' else 0 for each in data.soru3]
data.soru4=[1 if each=='Evet' else 0 for each in data.soru4]
data.soru5=[1 if each=='Evet' else 0 for each in data.soru5]
data.soru6=[1 if each=='Evet' else 0 for each in data.soru6]
data.soru7=[1 if each=='Evet' else 0 for each in data.soru7]
data.soru8=[1 if each=='Evet' else 0 for each in data.soru8]
data.soru9=[1 if each=='Evet' else 0 for each in data.soru9]
data.soru10=[1 if each=='Evet' else 0 for each in data.soru10]

# Mapping Education
educational_mapping= {'Lisans Üstü':5,"Lisans":4,"Ön Lisans":3,"Lise":2,'Ortaokul':1,"İlkokul":0}
data.Egitim = data.Egitim.map(educational_mapping)

# Mapping Political Choices
parti_mapping = {   'IYI PARTI':5,
                    "CHP":4,
                    "DIĞER":3,
                    "AKP":2,
                    "MHP":1,
                    "HDP":0}
data.parti = data.parti.map(parti_mapping)

# Mapping Ages
age_mapping = {'0-18':0,'18-30':1,'30-50':2,'50-60':3,'60+':4}
data.Yas = data.Yas.map(age_mapping)

# Mapping Bolge
bolge_mapping = {'Marmara':0,'Ege':1,'İç Anadolu':2,'Akdeniz':3,'Karadeniz':4,'Güneydoğu':5,'Doğu Anadolu':6}
data.Bolge = data.Bolge.map(bolge_mapping)
data.head(10)
```
